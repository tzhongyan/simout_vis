```python
with open('simout/simge00001.nml') as f:
    content = [l.strip() for l in f]
```

```python
# strip first line
content = content[1:]
```

```python
ge = {}
```

```python
# test_content = content[:100]
this_key = None
for line in content:
    eq_sp = line.split('=')
    nums = None
    if len(eq_sp) > 1:
        this_key = eq_sp[0].strip()
        ge[this_key] = []
        nums = eq_sp[1]
    else:
        nums = eq_sp[0]
    
    nums = nums.split(',')
    ge[this_key].extend([n.strip() for n in nums])
    
```

```python
for k, v in ge.items():
    ge[k] = [eval(x) for x in ge[k] if x ]
```

```python
for k, v in ge.items():
    print(k, len(ge[k]))
```
